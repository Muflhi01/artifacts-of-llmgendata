```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a right-skewed distribution using log-normal distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.lognormal(mu, sigma, 1000)

# Plotting the distribution
count, bins, ignored = plt.hist(s, 30, density=True, align='mid')

# Plot the distribution curve
x = np.linspace(min(bins), max(bins), 10000)
pdf = (np.exp(-(np.log(x) - mu)**2 / (2 * sigma**2))
       / (x * sigma * np.sqrt(2 * np.pi)))

plt.plot(x, pdf, linewidth=2, color='r')
plt.axis('tight')
plt.show()

# Return the generated numbers for inspection
s[:10] # displaying only the first 10 values

```

```python
# Adjusting the mean and standard deviation to generate a slightly right-skewed distribution
mu_slight, sigma_slight = 0, 0.5  # Increased variance for slight skew
s_slight_skew = np.random.lognormal(mu_slight, sigma_slight, 1000)

# Plotting the slightly right-skewed distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
count_slight, bins_slight, ignored_slight = plt.hist(s_slight_skew, 30, density=True, align='mid', color='orange', alpha=0.7)
x_slight = np.linspace(min(bins_slight), max(bins_slight), 10000)
pdf_slight = (np.exp(-(np.log(x_slight) - mu_slight)**2 / (2 * sigma_slight**2))
              / (x_slight * sigma_slight * np.sqrt(2 * np.pi)))
plt.plot(x_slight, pdf_slight, linewidth=2, color='r')
plt.title('Slightly Right-Skewed Distribution')
plt.axis('tight')

# Adjusting the parameters to generate a highly right-skewed distribution
mu_high, sigma_high = 0, 1  # Increased variance for high skew
s_high_skew = np.random.lognormal(mu_high, sigma_high, 1000)

# Plotting the highly right-skewed distribution
plt.subplot(1, 2, 2)
count_high, bins_high, ignored_high = plt.hist(s_high_skew, 30, density=True, align='mid', color='blue', alpha=0.7)
x_high = np.linspace(min(bins_high), max(bins_high), 10000)
pdf_high = (np.exp(-(np.log(x_high) - mu_high)**2 / (2 * sigma_high**2))
            / (x_high * sigma_high * np.sqrt(2 * np.pi)))
plt.plot(x_high, pdf_high, linewidth=2, color='r')
plt.title('Highly Right-Skewed Distribution')
plt.axis('tight')

plt.tight_layout()
plt.show()

# Return the first 10 generated numbers for both distributions for inspection
(s_slight_skew[:10], s_high_skew[:10])
```

```python
# To create various distributions, we will use the following strategies:
# - Left Skewed: Use skewnorm with a negative skewness parameter.
# - Normal: Use numpy's normal distribution function.
# - Slightly Right Skewed: Use lognormal with small sigma.
# - Highly Right Skewed: Use lognormal with larger sigma.
# - Very Highly Right Skewed: Use lognormal with even larger sigma.
import seaborn as sns
from scipy.stats import skewnorm, norm

# Set up the subplots
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

# Generate and plot each distribution
# Left Skewed
a = -10  # Negative value for left skew
data_left = skewnorm.rvs(a, size=1000)
sns.histplot(data_left, kde=True, color="skyblue", ax=axes[0], stat="density")
axes[0].set_title('Left Skewed')

# Normal Distribution
data_normal = np.random.normal(size=1000)
sns.histplot(data_normal, kde=True, color="skyblue", ax=axes[1], stat="density")
axes[1].set_title('Normal')

# Slightly Right Skewed
mu, sigma = 0, 0.5
data_slight_right = np.random.lognormal(mu, sigma, 1000)
sns.histplot(data_slight_right, kde=True, color="skyblue", ax=axes[2], stat="density")
axes[2].set_title('Slightly Right Skewed')

# Highly Right Skewed
mu, sigma = 0, 1
data_high_right = np.random.lognormal(mu, sigma, 1000)
sns.histplot(data_high_right, kde=True, color="skyblue", ax=axes[3], stat="density")
axes[3].set_title('Highly Right Skewed')

# Very Highly Right Skewed
mu, sigma = 0, 1.5
data_very_high_right = np.random.lognormal(mu, sigma, 1000)
sns.histplot(data_very_high_right, kde=True, color="skyblue", ax=axes[4], stat="density")
axes[4].set_title('Very Highly Right Skewed')

plt.tight_layout()
plt.show()
```
